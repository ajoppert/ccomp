```python
import numpy as np
import glob, os
import networkx as nx
import matplotlib.pyplot as plt 

#############################################
# Professores: Luerbio Faria e Diana Sasaki
# Aluno: Angelo Joppert
#############################################
'''
Funcionamento do programa:
Entrada: Dar opção ao usuário ao usuário
1. carregar um grafo, ou 2. criar um grafo, ou 3. sair.

1. Se ele desejar carregar o grafo,
a) Listar os grafos disponíveis na memória.
b) O usuário deverá dizer o nome do grafo,
c) O programa carregará o grafo e imprimirá a Matriz de adjacência e retornar para opções.

2 Se ele desejar criar um grafo.
a) O programa lerá o nome do grafo G=(V,E) para criar informando o usuário se o nome já foi usado, para pedir pelo nome novamente.
b) O programa lerá um Inteiro Positivo n.
c) O programa deverá receber do usuário as entradas da matriz triangular superior da matriz de adjacência M_{nxn} de G.
d) O programa deverá armazenar M na memória para carregamento posterior, imprimir M e retornar para opções .

O Programa deverá ser acompanhado de uma entrada (arquivo) G criada com a Opção 2.
'''

def listStoredGraphs():
  countsavedGraphs = len(glob.glob("*.npy"))
  if countsavedGraphs > 0:
    print("\nGrafo(s) em memória:")
    for file in glob.glob("*.npy"):
        print(file)
  return countsavedGraphs        

def upperTriangularToSymmetric(matrix, n):
    for i in range(1, n):
        for j in range(i):
            matrix[i, j] = matrix[j, i]

def showGraphFromAdjacentMatrix(adjacent, N):
    G = nx.DiGraph() 
    for i in range(N): 
        for j in range(N): 
            if adjacent[i][j] == 1: 
                G.add_edge(i,j)  
    nx.draw(G) 
    plt.show()                     

def printMatrix(matrix, N):   
    for i in range(N):
        for j in range(N):
            print(matrix[i][j], end = " ")
        print()
    print()

def fileNameExists(fileName):
  return os.path.isfile(fileName)

def loadMatrixElements(matrix, N):   
    # carrega elementos da matriz
    for i in range(N):  # linhas
        for j in range(N):  # colunas
            if i<=j:
               #Recebe, do usuário, elementos da matriz triangular superior e da DP
               matrix[i][j] = int(input())

def subMenuLoadGraph():
    countsavedGraphs = listStoredGraphs()
    if countsavedGraphs > 0:
      nomeGrafo = input("Informe o nome do grafo a ser carregado(sem extensão)!\n")
      data = np.load(nomeGrafo+'.npy')
      print(data)
      showGraphFromAdjacentMatrix(data, len(data))
    else:
      print("Não há grafos salvos. Selecione a opção 2 para criar um grafo!\n")

def subMenuCreateGraph():
    contador = 0
    while True:
      if contador == 0:
          nomeGrafo = input("Informe nome do grafo(sem extensão)!\n")
      else:
          nomeGrafo = input("O nome informado já foi utilizado! Tente outro nome para o grafo(sem extensão)!\n")
      contador = contador + 1
      if not fileNameExists(nomeGrafo+'.npy'):
        break

    N = int(input("Digite o valor de n.\n"))
    # inicializa a matriz com zeros
    matrix = np.zeros((N, N))
    print("Informe as entradas da matriz triangular superior da matriz de adjacência M_{nxn} de G:") 
    loadMatrixElements(matrix, N)
    upperTriangularToSymmetric(matrix, N)     
    np.save(nomeGrafo+'.npy', matrix)
    printMatrix(matrix, N)
    showGraphFromAdjacentMatrix(matrix, N)    

def menuprinc():
    opcao=int(input('''Escolha uma das opções:\n1 - Carregar um grafo\n2 - Criar um grafo\n3 - Sair\nEscolha:  '''))
    if opcao == 1:
        subMenuLoadGraph()
    elif opcao == 2:
        subMenuCreateGraph()
    elif opcao == 3:
        exit()
    else:
        print("A opção informada não está disponível, tente novamente!\n")
        menuprinc()
    return opcao    

#FUNÇÃO PRINCIPAL
loopControl = True
while (loopControl):
    if menuprinc() == 3:
        loopControl = False         
```
